### 4.1 Missing Values

````python
# =============================================================================
# 4.1 MISSING VALUE ANALYSIS
# =============================================================================

missing_rows = []

for name, df in datasets.items():
    missing_count = df.isnull().sum()
    missing_pct   = (missing_count / len(df) * 100).round(2)

    temp = pd.DataFrame({
        "Dataset"       : name,
        "Variable"      : missing_count.index,
        "Missing Count" : missing_count.values,
        "Missing (%)"   : missing_pct.values,
    })
    missing_rows.append(temp)

missing_summary = pd.concat(missing_rows, ignore_index=True)

has_missing = missing_summary[missing_summary["Missing Count"] > 0]
print(f"⚠️  Variables with missing values: {len(has_missing)}")
has_missing.sort_values("Missing (%)", ascending=False).head(20)
````

### 4.2 Duplicate Records

````python
# =============================================================================
# 4.2 DUPLICATE RECORD DETECTION
# =============================================================================

duplicate_report = pd.DataFrame([
    {"Dataset": name, "Duplicate Rows": df.duplicated().sum(), "Status": "✅ Clean" if df.duplicated().sum() == 0 else "⚠️  Duplicates found"}
    for name, df in datasets.items()
])

duplicate_report
````

### 4.3 Date Parsing

````python
# =============================================================================
# 4.3 DATE COLUMN PARSING
# =============================================================================

sanctions["date_implemented"] = pd.to_datetime(sanctions["date_implemented"], errors="coerce")

print("✅ Date columns parsed")
print(f"   Range: {sanctions['date_implemented'].min().date()} → {sanctions['date_implemented'].max().date()}")
````

### 4.4 Country Name Consistency

````python
# =============================================================================
# 4.4 COUNTRY NAME CONSISTENCY
# =============================================================================

trade_countries  = set(trade["exporter_country"])
budget_countries = set(budget["country"])
energy_countries = set(energy["country"])
macro_countries  = set(macro["country"])

consistency_report = pd.DataFrame({
    "Dataset"          : ["Trade", "Budget", "Energy", "Macro"],
    "Unique Countries" : [len(trade_countries), len(budget_countries), len(energy_countries), len(macro_countries)],
})

print(consistency_report.to_string(index=False))
print(f"\n📌 Common countries (all datasets): {len(trade_countries & budget_countries & energy_countries & macro_countries)}")
````

### 4.6 Validity Checks

````python
# =============================================================================
# 4.6 VALIDITY CHECKS — PERCENTAGE COLUMNS & NEGATIVE VALUES
# =============================================================================

# Percentage bounds check
for col in ["public_investment_pct", "private_investment_pct"]:
    invalid = budget[(budget[col] < 0) | (budget[col] > 100)]
    status  = "✅ Valid" if len(invalid) == 0 else f"⚠️  {len(invalid)} records out of [0,100]"
    print(f"{col}: {status}")

# Negative value check
print("\n── Negative Value Checks ──────────────────────────────────────────────")
for dataset_name, df in [("Budget", budget), ("Energy", energy), ("Macro", macro)]:
    numeric_cols = df.select_dtypes(include=np.number).columns
    neg_found    = (df[numeric_cols] < 0).sum()
    neg_found    = neg_found[neg_found > 0]

    if len(neg_found):
        print(f"\n⚠️  {dataset_name}:"); print(neg_found)
    else:
        print(f"✅ {dataset_name}: No unexpected negatives")
````

### 4.7 Data Quality Scorecard

````python
# =============================================================================
# 4.7 DATA QUALITY SCORECARD
# =============================================================================

scorecard = []
for name, df in datasets.items():
    completeness = (1 - df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
    uniqueness   = (1 - df.duplicated().sum() / len(df)) * 100
    scorecard.append({
        "Dataset"          : name,
        "Completeness (%)" : round(completeness, 2),
        "Uniqueness (%)"   : round(uniqueness, 2),
        "Status"           : "🟢 Good" if completeness >= 90 and uniqueness == 100 else "🟡 Acceptable",
    })

scorecard_df = pd.DataFrame(scorecard)
scorecard_df
````

````python
# =============================================================================
# DATA QUALITY SCORECARD — VISUALIZATION
# =============================================================================

plot_data = scorecard_df.melt(
    id_vars="Dataset",
    value_vars=["Completeness (%)", "Uniqueness (%)"],
    var_name="Metric", value_name="Score"
)

fig, ax = plt.subplots(figsize=(12, 5))
sns.barplot(data=plot_data, x="Dataset", y="Score", hue="Metric", ax=ax)
ax.set_title("Data Quality Scorecard by Dataset", fontweight="bold")
ax.set_ylim(0, 105)
ax.axhline(90, linestyle="--", color="red", alpha=0.5, label="90% threshold")
ax.legend()
plt.tight_layout()
plt.show()
````

## 7. Feature Engineering

Raw variables are transformed into **12 strategic risk indicators** across five categories:

| Category | Feature | Formula |
|----------|---------|---------|
| Supply Chain | Supplier Dependency Index (SDI) | Max supplier share per importer |
| Supply Chain | Trade Concentration Index (HHI) | Σ supplier_share² |
| Geopolitical | Geopolitical Exposure Index | 0.4 × exporter_risk + 0.4 × importer_risk + 0.2 × restriction |
| Geopolitical | Sanction Pressure Score | 0.4 × count + 0.3 × severity + 0.3 × impact |
| AI Infrastructure | AI Investment Intensity | investment / GDP |
| AI Infrastructure | AI Sovereignty Score | sovereignty_index × investment |
| Energy | Energy Constraint Index | pct_grid_used × stress_index |
| Energy | Renewable Advantage Score | renewable_pct / 100 |
| Corporate | Corporate Exposure Score | geo_exposure × (1 + sanction_risk) |
| Corporate | Innovation Capacity Score | R&D / revenue |
| Composite | Trade Route Risk Score | 0.5 × geo_exposure + 0.5 × sanction_pressure |
| Composite | AI Readiness Score | 0.35 × investment + 0.30 × sovereignty + 0.20 × grid + 0.15 × tech_gdp |

---
### 7.1 Supplier Dependency Index (SDI)

````python
# =============================================================================
# 7.1 SUPPLIER DEPENDENCY INDEX (SDI)
# Measures the maximum supplier share per importing country.
# Higher SDI = higher concentration risk.
# =============================================================================

supplier_trade = (
    master_df
    .groupby(["importer_country", "primary_supplier"])["trade_value_usd_millions"]
    .sum().reset_index()
)
````

````python
country_trade = (

    master_df

    .groupby("importer_country")

    ["trade_value_usd_millions"]

    .sum()

    .reset_index()

    .rename(
        columns={
            "trade_value_usd_millions":
            "country_trade"
        }
    )

)
````

````python
supplier_trade = supplier_trade.merge(

    country_trade,

    on="importer_country"

)

supplier_trade["supplier_share"] = (

    supplier_trade["trade_value_usd_millions"]

    /

    supplier_trade["country_trade"]

)
````

````python
sdi = (

    supplier_trade

    .groupby("importer_country")

    ["supplier_share"]

    .max()

    .reset_index()

)

sdi.columns = [

    "importer_country",
    "supplier_dependency_index"

]

sdi.head()
````

````python
master_df = master_df.merge(

    sdi,

    on="importer_country",

    how="left"

)
````

### 7.2 Trade Concentration Index (HHI)

````python
# =============================================================================
# 7.2 HERFINDAHL-HIRSCHMAN INDEX (HHI)
# Measures market concentration. Higher HHI = greater dependency risk.
# =============================================================================

hhi = (
    supplier_trade
    .groupby("importer_country")["supplier_share"]
    .apply(lambda x: np.sum(x ** 2))
    .reset_index()
)

hhi.columns = ["importer_country", "trade_concentration_index"]
````

````python
master_df = master_df.merge(

    hhi,

    on="importer_country",

    how="left"

)
````

### 7.3 Geopolitical & Sanction Features

````python
master_df["geopolitical_exposure_index"] = (

    master_df["geo_risk_score_exporter"]

    * 0.4

    +

    master_df["geo_risk_score_importer"]

    * 0.4

    +

    master_df["restriction_factor"]

    * 0.2

)
````

````python
master_df["sanction_pressure_score"] = (

    master_df["sanctions_count"]

    * 0.4

    +

    master_df["avg_sanction_severity"]

    * 0.3

    +

    master_df["economic_impact_usd_bn"]

    * 0.3

)
````

### 7.4 AI Infrastructure Features

````python
master_df["ai_investment_intensity"] = (

    master_df["total_ai_investment_usd_bn"]

    /

    (master_df["gdp_usd_bn"] + 1)

)
````

````python
master_df["ai_sovereignty_score"] = (

    master_df["ai_sovereignty_index"]

    *

    master_df["total_ai_investment_usd_bn"]

)
````

### 7.5 Energy Features

````python
master_df["energy_constraint_index"] = (

    master_df["pct_grid_used_by_ai"]

    *

    master_df["ai_grid_stress_index"]

)
````

````python
master_df["renewable_advantage_score"] = (

    master_df["renewable_energy_pct"]

    /

    100

)
````

### 7.6 Corporate Features

````python
master_df["corporate_exposure_score"] = (

    master_df["avg_geopolitical_exposure"]

    *

    (1 + master_df["sanction_risk_ratio"])

)
````

````python
master_df["innovation_capacity_score"] = (

    master_df["total_rd_usd_bn"]

    /

    (master_df["total_revenue_usd_bn"] + 1)

)
````

### 7.7 Trade Route & AI Readiness Scores

````python
master_df["trade_route_risk_score"] = (

    master_df["geopolitical_exposure_index"]

    * 0.5

    +

    master_df["sanction_pressure_score"]

    * 0.5

)
````

````python
master_df["ai_readiness_score"] = (

    master_df["total_ai_investment_usd_bn"] * 0.35

    +

    master_df["ai_sovereignty_index"] * 0.30

    +

    master_df["total_grid_capacity_gw"] * 0.20

    +

    master_df["tech_sector_pct_gdp"] * 0.15

)
````

### 7.8 Feature Summary & Correlation

````python
engineered_features = [

    "supplier_dependency_index",

    "trade_concentration_index",

    "geopolitical_exposure_index",

    "sanction_pressure_score",

    "ai_investment_intensity",

    "ai_sovereignty_score",

    "energy_constraint_index",

    "renewable_advantage_score",

    "corporate_exposure_score",

    "innovation_capacity_score",

    "trade_route_risk_score",

    "ai_readiness_score"

]

master_df[engineered_features].describe()
````

## 8. Risk Intelligence Framework

### Architecture

```
Supply Chain Risk Score   (35%)  ──┐
Geopolitical Risk Score   (30%)  ──┤
Infrastructure Risk Score (20%)  ──┤──► Composite Risk Score ──► Risk Tier
Corporate Risk Score      (15%)  ──┘
```

All sub-indicators are **Min-Max normalized** to [0, 1] before weighting.

---
### 8.1 Feature Normalization

````python
risk_df = master_df.copy()
````

````python
# =============================================================================
# 8.1 RISK FEATURES TO NORMALIZE
# =============================================================================

risk_features = [
    "supplier_dependency_index",
    "trade_concentration_index",
    "geopolitical_exposure_index",
    "sanction_pressure_score",
    "energy_constraint_index",
    "corporate_exposure_score",
    "trade_route_risk_score",
]
````

````python
# =============================================================================
# MIN-MAX SCALING
# =============================================================================

scaler = MinMaxScaler()

risk_df[[f"{col}_scaled" for col in risk_features]] = scaler.fit_transform(
    risk_df[risk_features]
)

print("✅ All risk features normalized to [0, 1]")
````

### 8.2 Component Risk Scores

````python
risk_df["supply_chain_risk_score"] = (

    risk_df["supplier_dependency_index_scaled"]

    * 0.50

    +

    risk_df["trade_concentration_index_scaled"]

    * 0.50

)
````

````python
risk_df["geopolitical_risk_score_final"] = (

    risk_df["geopolitical_exposure_index_scaled"]

    * 0.60

    +

    risk_df["sanction_pressure_score_scaled"]

    * 0.40

)
````

````python
risk_df["infrastructure_risk_score"] = (

    risk_df["energy_constraint_index_scaled"]

)
````

````python
risk_df["corporate_risk_score"] = (

    risk_df["corporate_exposure_score_scaled"]

)
````

### 8.3 Composite Risk Score

````python
# =============================================================================
# 8.3 COMPOSITE RISK SCORE
# Weights: Supply Chain 35% | Geopolitical 30% | Infrastructure 20% | Corporate 15%
# =============================================================================

risk_df["composite_risk_score"] = (
    risk_df["supply_chain_risk_score"]         * 0.35
    + risk_df["geopolitical_risk_score_final"] * 0.30
    + risk_df["infrastructure_risk_score"]     * 0.20
    + risk_df["corporate_risk_score"]          * 0.15
)
````

````python
risk_df["composite_risk_score"].describe()
````

````python
plt.figure(figsize=(10,6))

sns.histplot(
    risk_df["composite_risk_score"],
    bins=30,
    kde=True
)

plt.title(
    "Composite Risk Score Distribution"
)

plt.show()
````

### 8.4 Risk Tier Segmentation

Countries are categorized into four tiers using quartile thresholds:

| Tier | Quartile | Recommended Action |
|------|----------|-------------------|
| 🟢 Low | Q1 | Routine monitoring |
| 🟡 Medium | Q2 | Periodic review |
| 🔴 High | Q3 | Active mitigation |
| ⚫ Critical | Q4 | Immediate intervention |

````python
risk_df["risk_tier"] = pd.qcut(

    risk_df["composite_risk_score"],

    q=4,

    labels=[

        "Low",

        "Medium",

        "High",

        "Critical"

    ]

)
````

````python
plt.figure(figsize=(8,5))

sns.countplot(
    data=risk_df,
    x="risk_tier"
)

plt.title(
    "Risk Tier Distribution"
)

plt.show()
````

### 8.5 High-Risk Transaction Flagging

````python
# =============================================================================
# 8.6 HIGH-RISK TRANSACTION FLAG (Top 20th Percentile)
# =============================================================================

risk_threshold = risk_df["composite_risk_score"].quantile(0.80)

risk_df["high_risk_transaction"] = np.where(
    risk_df["composite_risk_score"] >= risk_threshold, 1, 0
)

print(f"Risk threshold (80th pct): {risk_threshold:.4f}")
````
